```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df0=pd.read_csv("Table1.csv")
df1=pd.read_csv("Table2.csv")
df2=pd.read_csv("Table3.csv")
df3=pd.read_csv("Table4.csv")
df4=pd.read_csv("Table5.csv")
df5=pd.read_csv("Table6.csv")
```

```python
# gca stands for 'get current axis'
ax = plt.gca()

df0.plot(kind='line',x='F1',y='F2',ax=ax,grid=True,figsize=(10,6),title='Dipole moments and Stokes frequencies at incident radiation of frequency 35.38 s-1',xlabel='Dipole Moments',ylabel='Modified  Stokes  frequencies')
df0.plot(kind='line',x='F1',y='F3', color='red', ax=ax,grid=True,figsize=(10,6),title='Dipole moments and Stokes frequencies at incident radiation of frequency 35.38 s-1',xlabel='Dipole Moments',ylabel='Modified  Stokes  frequencies')
```

```python
# sample points 
X = list(df0['F1'])
Y = list(df0['F2'])

# solve for a and b
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b

# solution
a, b = best_fit(X, Y)



# plot points and fit line
import matplotlib.pyplot as plt
plt.scatter(X, Y)
yfit = [a + b * xi for xi in X]
plt.plot(X, yfit, label='w1+w2+wD')

# sample points 
X1= list(df0['F1'])
Y1= list(df0['F3'])

# solve for a and b
def best_fit(X1, Y1):

    x1bar = sum(X1)/len(X1)
    y1bar = sum(Y1)/len(Y1)
    n = len(X1) # or len(Y1)

    numer1 = sum([xj*yj for xj,yj in zip(X1, Y1)]) - n * x1bar * y1bar
    denum1 = sum([xj**2 for xj in X1]) - n * x1bar**2

    b1 = numer1 / denum1
    a1 = y1bar - b1 * x1bar

    print('best fit line:\ny1 = {:.2f} + {:.2f}x1'.format(a1, b1))

    return a1, b1

# solution
a1, b1 = best_fit(X1, Y1)


# plot points and fit line
import matplotlib.pyplot as plt
plt.scatter(X1, Y1)
y1fit = [a1 + b1 * xj for xj in X1]
plt.plot(X1, y1fit, label='w1+w2-wD')





plt.xlabel('DIPOLE MOMENTS')
# Set the y axis label of the current axis.
plt.ylabel('MODIFIED STOKES FREQUENCIES')
# Set a title of the current axes.
plt.title('GRAPH OF INCIDENT RADIIATION OF FREQUENCY 35.38 s-1')
# show a legend on the plot
plt.legend()

plt.grid(color = 'black', linestyle = '--', linewidth = 0.5)
# Display a figure.
plt.show()
```
